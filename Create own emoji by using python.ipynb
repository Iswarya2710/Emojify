```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
fig,ax=plt.subplots()

face_color='#ffd700'
edge_color='#000000'

head_radius=1.0
head=patches.Circle((0,0),radius=head_radius,facecolor=face_color,edgecolor=edge_color)
ax.add_patch(head)

eye_radius=0.2
eye_offsite=0.35
left_eye=patches.Circle((-eye_offsite,eye_offsite),radius=eye_radius,facecolor=edge_color,edgecolor=edge_color)
right_eye=patches.Circle((eye_offsite,eye_offsite),radius=eye_radius,facecolor=edge_color,edgecolor=edge_color)
ax.add_patch(left_eye)
ax.add_patch(right_eye)

mouth_width=0.5
mouth_height=0.3
mouth=patches.Arc((0,-0.2),width=mouth_width,height=mouth_height,theta1=45,theta2=135,facecolor=face_color,edgecolor=edge_color,lw=2)
ax.add_patch(mouth)

ax.set_aspect('equal')
ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_smiley_face(x, y, face_color, eye_color, mouth_color, expression):
    fig, ax = plt.subplots()
    
    head_radius = 1.0
    head = patches.Circle((x, y), radius=head_radius, facecolor=face_color, edgecolor='black')
    ax.add_patch(head)
    
    eye_radius = 0.15
    left_eye = patches.Circle((x - 0.3, y + 0.3), radius=eye_radius, facecolor=eye_color, edgecolor='black')
    right_eye = patches.Circle((x + 0.3, y + 0.3), radius=eye_radius, facecolor=eye_color, edgecolor='black')
    ax.add_patch(left_eye)
    ax.add_patch(right_eye)
    
    if expression == 'sad':
        mouth = patches.Arc((x, y - 0.2), width=0.5, height=0.3, theta1=0, theta2=180, facecolor=mouth_color, edgecolor='black',lw=2)
        print("sad")
    elif expression == 'neutral':
        mouth = patches.Arc((x, y - 0.4), width=0.5, height=0.3, theta1=45, theta2=135, facecolor=mouth_color, edgecolor='black',lw=2)
        print("neutral")
    elif expression == 'smile':
        mouth = patches.Arc((x, y - 0.4), width=0.5, height=0.3, theta1=180, theta2=360, facecolor=mouth_color, edgecolor='black',lw=2)
        print("smile")
    else:
        raise ValueError('Invalid expression')
    
    ax.add_patch(mouth)
    
    ax.set_aspect('equal')
    ax.set_xlim(x - head_radius - 0.5, x + head_radius + 0.5)
    ax.set_ylim(y - head_radius - 0.5, y + head_radius + 0.5)
    
    ax.axis('off')
    
    plt.show()

draw_smiley_face(0, 0, 'yellow', 'black', 'red', 'smile')
draw_smiley_face(3, 0, 'yellow', 'black', 'blue', 'neutral')
draw_smiley_face(6, 0, 'yellow', 'black', 'green', 'sad')
```
